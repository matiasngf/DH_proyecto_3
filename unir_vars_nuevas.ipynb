```python
X = pd.get_dummies(featuresNuevas.drop(['user_id', 'user_id.1', 'bot', 'Cat', 'tweetsCount', 'semanas'], axis=1))
atributos = X.columns
X = preprocessing.scale(X, axis = 0)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(featuresNuevas['bot'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.tree import DecisionTreeClassifier
treeclf = DecisionTreeClassifier(max_depth=5, random_state=1)
treeclf.fit(X_train, y_train)
pd.DataFrame({'Atributo':atributos,
              'importancia':treeclf.feature_importances_}).sort_values('importancia',
                                                                      ascending=False).head(20)
```

```python
from sklearn.metrics import confusion_matrix
conf = confusion_matrix(y_test, treeclf.predict(X_test))


predicted_cols = ['pred_'+str(c) for c in le.classes_]


pd.DataFrame(conf, index = le.classes_, columns = predicted_cols)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
param_trees = {'n_estimators': [50, 100, 200], 
               'max_features': [1, 5, 8, 10, 21], 
               'max_depth': [5, 20, 50, 70, 100], 
               'min_samples_leaf':[1, 5, 8, 10, 50]}
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(class_weight='balanced')
kf = StratifiedKFold(n_splits=3, shuffle=True)
grid_search_rf = GridSearchCV(rf, param_grid=param_trees, cv=kf, verbose=1, n_jobs=-1)
```

```python
grid_search_rf.fit(X_train, y_train)
```

```python
model = grid_search_rf.best_estimator_
grid_search_rf.best_score_
```

```python
# Predecir valores de test
y_pred = model.predict(X_test)
# Calculamos el accuracy general del modelo
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
conf = confusion_matrix(y_test, y_pred)


predicted_cols = ['pred_'+str(c) for c in le.classes_]


pd.DataFrame(conf, index = le.classes_, columns = predicted_cols)
```

```python
from sklearn.preprocessing import StandardScaler

dfSin3 = featuresNuevas[featuresNuevas.Cat!='SocialSpambot3']

X = pd.get_dummies(dfSin3.drop(['user_id', 'user_id.1', 'bot', 'Cat', 'tweetsCount', 'semanas'], axis=1))
atributos = X.columns
# X = preprocessing.scale(X, axis = 0)

le = LabelEncoder()
y = le.fit_transform(dfSin3['bot'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
param_trees = {'n_estimators': [200, 300], 
               'max_features': [21, 30, 40], 
               'max_depth': [100], 
               'min_samples_leaf':[1, 5, 8, 10, 50]}

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(class_weight='balanced')
kf = StratifiedKFold(n_splits=3, shuffle=True)
grid_search_rf = GridSearchCV(rf, param_grid=param_trees, cv=kf, verbose=1, n_jobs=-1)


X_train
```

```python
grid_search_rf.fit(X_train, y_train)
```

```python
model = grid_search_rf.best_estimator_
print('best score', grid_search_rf.best_score_)

# Predecir valores de test
y_pred = model.predict(X_test)
# Calculamos el accuracy general del modelo
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
